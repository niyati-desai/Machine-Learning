Purpose: Pivot tables are used for summarizing and aggregating data, while crosstabs are used for computing frequency tables.

Aggregation: Pivot tables can apply aggregate functions(e.g., mean, sum) to the values, while crosstabs simply count occurrences.

Inputs: Pivot tables require specifying values, indexes and columns, while crosstabs only require the factors to cross-tablulate.

Both pivot tables and crosstabs are valueable tools for exploring and summerizing data, and the choices between them depends on the specific analysis you're conducting.

Feature Engineering

```python
#Create new features
#total stats without HP
df['Total_without_HP'] = df['Total'] - df['HP']
```

```python
df.head()
```

```python
#Normalizing Features
from sklearn.preprocessing import MinMaxScaler

#Normalize attack, defense and speed
scalar= MinMaxScaler()
df[['Attack_scaled', 'Defense_scaled', 'Speed_scaled']] = scalar.fit_transform(df[['Attack', 'Defense', 'Speed']])
```

```python
df.head()
```

```python
#binning
bins=[0,50,100,150,200]
labels=['Very Slow', 'Slow', 'Fast', 'Very Fast']
df['Speed_category'] = pd.cut(df['Speed'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
fast_pokemons = df[df['Speed_category'] == 'Fast']
print(fast_pokemons)
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Legendary_Encoded'] = le.fit_transform(df['Legendary'])
```
